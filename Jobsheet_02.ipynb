# 4. Operasi matriks dan vektor dengan python

a. Membuat matriks di python

```python
import numpy as np

A = np.array([[80, 30, 25],
              [50, 20, 10]])

B = np.array([[30, 60, 15],
              [10, 30, 30]])

hasil = A + B
print("Hasil A + B:")
print(hasil)
```

b. Melakukan operasi dot product

```python
import numpy as np

A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

hasil = np.dot(A, B)
print("Hasil A . B:", hasil)
```

c. Melakukan operasi cross product

```python
import numpy as np

A = np.array([10, 20, 15])
B = np.array([16, 67, 35])

hasil = np.cross(A, B)
print("Hasil A x B:", hasil)
```

```python
import numpy as np

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])
C = np.array([7, 8, 9])

# A x B (cross product)
hasil_cross = np.cross(A, B)
print("A x B =", hasil_cross)

# B . C (dot product)
hasil_dot = np.dot(B, C)
print("B . C =", hasil_dot)
```

```python
import math

# 6!
faktorial_6 = math.factorial(6)
print("6! =", faktorial_6)

# 8! / 5!
hasil = math.factorial(8) / math.factorial(5)
print("8! / 5! =", hasil)
```

```python
import numpy as np

A = np.array([5, 3, 7])
B = np.array([12, -3, 1])

# Panjang vektor
panjang_A = np.linalg.norm(A)
panjang_B = np.linalg.norm(B)

print("Panjang A =", panjang_A)
print("Panjang B =", panjang_B)

# Dot product
dot = np.dot(A, B)
print("A . B =", dot)

# Cross product
cross = np.cross(A, B)
print("A x B =", cross)
```
