# LATIHAN

# 1. Central Tendency

a. Mendapatkan Nilai Mean

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

mean = np.mean(score)
print("Mean =", mean)
```

b. Mendapatkan Nilai Media

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

median = np.median(score)
print("Median =", median)
```

c. Mendapatkan Nilai Modus/Mode

```python
from scipy import stats

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

mode = stats.mode(score, keepdims=True)
print("Mode =", mode)
```

# 2. Spread

a. Mendapatkan Nilai Range

```python
import numpy as np

data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

minimum = np.min(data)
maximum = np.max(data)

range_nilai = maximum - minimum
print("Range =", range_nilai)
```

b. Mendapatkan Nilai Percentile

```python
import numpy as np

data = [90, 80, 70, 65, 110, 80, 83, 67, 94]

p90 = np.percentile(data, 90)
print("Percentile 90 =", p90)
```

c. Mendapatkan Nilai Quartile dan Interquartile Range

```python
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
```

d. Mendapatkan Nilai Variance dan Standart Deviation

```python
import numpy as np

data = [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print("Mean =", mean)
print("Variance =", var)
print("Standard Deviation =", std)
```

a. Urutkan data dari yang terkecil sampai data yang terbesar

```python
data = [50,54,62,50,52,59,61,63,65,10,53,63,65,50,59,62,50,51,57,60,63,65,65,53,99]

data_sorted = sorted(data)
print(data_sorted)
```

b. Identifikasi nilai first quartile (Q1) dan third quartile (Q3), dan IQR

```python
import numpy as np

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
```

c. c.	Tentukan nilai upper bound dan lower bound

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
```

d. Tentukan nilai ujian yang menjadi outlier berdasarkan data hasil ujian diatas

```python
outlier = [x for x in data if x < lower_bound or x > upper_bound]
print("Outlier =", outlier)
```
